```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def get_tmdb_link(letterboxd_url):
    """Extract TMDB link from a Letterboxd movie page"""
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }
        
        response = requests.get(letterboxd_url.strip(), headers=headers)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Find the TMDB link
            tmdb_link = soup.find('a', {'data-track-action': 'TMDB'})
            
            if tmdb_link:
                return tmdb_link['href']
            else:
                return "Not found"
        else:
            return f"Error: {response.status_code}"
            
    except Exception as e:
        return f"Error: {str(e)}"

```

```python
def process_and_print_tmdb_urls(csv_file):
    """Process URIs and print only TMDB URLs"""
    
    # Load CSV
    df = pd.read_csv(csv_file, quotechar='"', skipinitialspace=True, on_bad_lines='skip')
    
    print("TMDB URLs:\n" + "="*50)
    
    for index, row in df.iterrows():
        letterboxd_url = row['Letterboxd URI']
        tmdb_url = get_tmdb_link(letterboxd_url)
        
        # Only print valid TMDB URLs
        if tmdb_url.startswith('https://'):
            print(tmdb_url)
        
        time.sleep(1)
    
    print("="*5)

# Usage
```

```python
def search_tmdb(movie_name, year):
    url = 'https://api.themoviedb.org/3/search/movie'
    headers = {'Authorization': f'Bearer {ACCESS_TOKEN}'}
    params = {'query': movie_name, 'year': year}
    
    response = requests.get(url, headers=headers, params=params)
    data = response.json()
    
    if data['results']:
        movie_id = data['results'][0]['id']
        return f'https://www.themoviedb.org/movie/{movie_id}/'
    return None

```

```python
def url_to_api_endpoint(tmdb_url):
    """Convert TMDB webpage URL to API endpoint"""
    movie_id = tmdb_url.split('/movie/')[1].rstrip('/')
    return f'https://api.themoviedb.org/3/movie/{movie_id}'

```

```python
tmdb_uri=[]
for url in tmdb_sample:
    api_url = url_to_api_endpoint(url)
    tmdb_uri.append(api_url)  # Output: https://api.themoviedb.org/3/movie/11
tmdb_uri
```

```python
response_payload=[]
for uri in tmdb_uri:
    response = requests.get(uri, headers=headers)
    response_payload.append(response.json())


response_payload[1]["origin_country"][0], response_payload[1]["original_language"], response_payload[1]["genres"][0]['name']

# response_payload


```
